```python
import pandas as pd

credit = pd.read_csv(
    "Dataset/credit_record.csv",
    header=None,
    names=["raw"]
)

credit = credit["raw"].str.replace('"', "", regex=False).str.split(",", expand=True)
credit.columns = ["ID", "MONTHS_BALANCE", "STATUS"]

credit.head()
```

```python
credit.columns
```

```python
credit["ID"] = credit["ID"].astype("int64")
credit["MONTHS_BALANCE"] = credit["MONTHS_BALANCE"].astype("int64")
```

```python
credit.head()
```

```python
credit = credit.iloc[1:].reset_index(drop=True)
```

```python
credit["ID"] = credit["ID"].astype("int64")
credit["MONTHS_BALANCE"] = credit["MONTHS_BALANCE"].astype("int64")
```

```python
credit.head()
credit.dtypes
```

```python
status_map = {
    'C': 1,
    'X': 1,
    '0': 0,
    '1': 1,
    '2': 1,
    '3': 1,
    '4': 1,
    '5': 1
}

credit["TARGET"] = credit["STATUS"].map(status_map)
```

```python
credit.head()
```

```python
credit_target = credit.groupby("ID")["TARGET"].max().reset_index()
```

```python
credit_target.head()
```

```python
data = application.merge(credit_target, on="ID", how="inner")
```

```python
data.head()
data.shape
```

```python
data.isnull().sum()
```

```python
data = data.drop_duplicates()
```

```python
data["AGE"] = (-data["DAYS_BIRTH"] / 365).astype(int)
```

```python
import numpy as np

data["DAYS_EMPLOYED"] = data["DAYS_EMPLOYED"].replace(365243, np.nan)
data["YEARS_EMPLOYED"] = (-data["DAYS_EMPLOYED"] / 365)
```

```python
data.drop(columns=["DAYS_BIRTH", "DAYS_EMPLOYED"], inplace=True)
```

```python
binary_cols = [
    "CODE_GENDER",
    "FLAG_OWN_CAR",
    "FLAG_OWN_REALTY"
]

for col in binary_cols:
    print(col, data[col].unique())
```

```python
data["OCCUPATION_TYPE"] = data["OCCUPATION_TYPE"].fillna("Unknown")
data["YEARS_EMPLOYED"] = data["YEARS_EMPLOYED"].fillna(data["YEARS_EMPLOYED"].median())
```

```python
data["CODE_GENDER"] = data["CODE_GENDER"].map({"M": 1, "F": 0})
data["FLAG_OWN_CAR"] = data["FLAG_OWN_CAR"].map({"Y": 1, "N": 0})
data["FLAG_OWN_REALTY"] = data["FLAG_OWN_REALTY"].map({"Y": 1, "N": 0})
```

```python
categorical_cols = [
    "NAME_INCOME_TYPE",
    "NAME_EDUCATION_TYPE",
    "NAME_FAMILY_STATUS",
    "NAME_HOUSING_TYPE",
    "OCCUPATION_TYPE"
]

data = pd.get_dummies(
    data,
    columns=categorical_cols,
    drop_first=True
)
```

```python
data.to_csv("Dataset/feature_engineered_data.csv", index=False)
```
