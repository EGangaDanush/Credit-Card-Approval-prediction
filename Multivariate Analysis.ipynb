# Multivariate Analysis

```python
# Income vs Age
plt.scatter(application_df["DAYS_BIRTH"],
            application_df["AMT_INCOME_TOTAL"])

plt.xlabel("Days Birth")
plt.ylabel("Income")
plt.title("Income vs Age")
plt.show()
```

```python
# Gender vs Income
application_df.boxplot(column="AMT_INCOME_TOTAL",
                       by="CODE_GENDER",
                       figsize=(6,5))
plt.show()
```

```python
# eduction vs Income
import seaborn as sns

sns.boxplot(x="NAME_EDUCATION_TYPE",
            y="AMT_INCOME_TOTAL",
            data=application_df)

plt.xticks(rotation=45)
plt.show()
```

```python
# Correlation heatmap
numeric = application_df.select_dtypes(include="number")

import seaborn as sns

plt.figure(figsize=(10,8))
sns.heatmap(numeric.corr(),
            annot=True,
            cmap="coolwarm")

plt.show()
```
